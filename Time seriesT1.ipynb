```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")

```

```python
df.info()
```

```python
df_clean = df.dropna(subset=["Order Date"])
```

```python
df_clean.info()
```

```python
import matplotlib.pyplot as plt

daily_sales = df_clean.groupby("Order Date")["Sales"].sum()

plt.figure(figsize=(10,4))
plt.plot(daily_sales)
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df_clean.resample("M", on="Order Date")["Sales"].sum()

plt.figure(figsize=(8,4))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
category_sales = (
    df_clean.groupby("Category")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(6,4))
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
category_sales.plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Category-wise Sales Share")
plt.tight_layout()
plt.show()
```
